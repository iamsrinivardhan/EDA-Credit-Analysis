```python
def subplot_Seaborn_CountPlot(column_name, figsize_x, figsize_y,rot):
    #fig,ax = plt.subplots(1,2, sharey = True, figsize=[figsize_x,figsize_y])
    plt.figure(figsize=[figsize_x,figsize_y])
    plt.subplot(1,2,1)
    plt.title('Non-Defaulter')
    sns.countplot(T0[column_name], palette='Set2')
    plt.xticks(rotation = rot)
    plt.subplot(1,2,2)
    plt.title('Defaulter')
    sns.countplot(T1[column_name], palette='Set2')
    plt.xticks(rotation = rot)
    plt.show()
```
